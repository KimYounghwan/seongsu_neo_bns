```python
# 평균계산
# 위치매개변수 가변매개변수: tuple타입으로 전달
def get_avg( *nums ):
    print(type(nums))
    return sum(nums) / len(nums)

get_avg( 70,80,90,)
get_avg( 70,80,90,70,60)

```

```python
# 키워드 매개변수 가변매개변수: 사전타입
def get_avg2( **nums ):
#     print(type(nums),nums)
    return sum(list(nums.values())) / len(nums.values())

print(get_avg2(영어=70,수학=80,국어=90))
print(get_avg2(영어=70,수학=80,과학=80))
```

```python
# 변수 읽을때엔 로컬검색 후 전역검색
# 변수 쓸때엔 로컬에 쓰기
count = 0
def inc():
    global count # count변수는 전역변수다 선언
    count = count + 5
    print("count=",count)

inc()
print("전역 count=",count)
inc
```

```python
def pv(a):
    b = 0
    b += a
    def get_b():
        return b
    def set_b(c):
        nonlocal b
        b += c # 누적
    
    return (get_b, set_b)

gb,gc = pv(5)
gb()
gc(9)
gb()
```
