```python
# 데이터 로딩, 데이터 전처리, 통계, 시각화
import numpy as np
```

```python
# ndarray 생성
arr = [1,2,3]
nd1 = np.array( arr )
display(arr, nd1)
```

```python
# numpy 인덱싱, 슬라이싱
data2d = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]
nd2 = np.array( data2d )
nd2
```

```python
print(data2d[1][0] ) # 4
print(nd2[1][0] ) # 4
print(nd2[1 ,0] ) # 4
```

```python
# 슬라이싱
print( nd2[ : , 0] ) # 모든행의 0번 칼럼들 추출
print( nd2[ : , 0:2] ) # 모든행의 0,1번 칼럼들 추출
# 팬시 인덱싱( 띄엄띄엄 추출 )
print( nd2[ : , [0,2]] ) # 모든행의 0,2번 칼럼들 추출
# 조건 인덱싱
bidx = nd1 > 2 # [False, False,  True]
nd1[ bidx ] # True에 해당하는 값만 추출 -> [3]
nd1[ [True, False, True] ] # [1, 3]
```

```python
# 2차원 배열
bidx = nd2 % 2 == 0 #짝수 검사
nd2[ bidx ]
```

```python
# range(시작,끝[,증가치-정수])  시작이상, 끝미만의 연속적인 정수 생성
# np.arange(시작,끝[,증가치-실수]) 시작이상, 끝미만의 연속적인 실수 생성
list(range(0,5))
np.arange(0, 0.5, 0.1)
```

```python
# np.reshape(arr, size=(row,col))
# arr.reshape(arr, size=(row,col))
nd3 = np.arange(12).reshape(4,3)
nd3.reshape(6,2)
# -1 길이 -> 자동계산.
nd3.reshape(2,3,2) # 면, 행, 열
```

```python
# 브로드캐스트 연산
# 배열 끼리 연산 : 같은 위치끼리 연산
a = np.array([1,2,3])
b = np.array([4,5,6])
a+b
```

```python
# 기술통계함수(description )
# sum(), mean()-평균, min(), max(), var(), std(), quantile()-백분위수
# comsum()-누적합 comprod()-누적곱
# axis = 0 ->행데이터를 읽어 열별 통계 계산
# axis = 1 ->열데이터를 읽어 행별 통계 계산
# axis = None 전체 통계
nd4 = np.arange(6).reshape(2,3)
print(nd4)
# 칼럼별 평균
print( nd4.mean() ) #전체평균
print( nd4.mean(axis=0) ) #칼럼평균
print( nd4.mean(axis=1) ) #행평균
```

```python
# 삭제하기
# df.drop( 인덱스 , axis=0/1)  axis=0 행삭제, axis=1 열삭제
# drop()은 삭제된 복사본 반환
# drop( inplace = True)  원본 삭제
apt_trading.drop("번지",axis=1)
apt_trading.drop(columns="번지")
apt_trading.drop(columns=["번지","부번","본번"])
```

```python
# 칼럼명 변경
apt_trading.rename(columns={"전용면적(㎡)":"전용면적"} , inplace=True)
```

```python
# 전용면적 / 3.03 -> 평 저장
apt_trading["pyung"] = round(apt_trading['전용면적'] / 3.03, 2)
apt_trading["계약년"] = apt_trading["계약년월"] // 100
apt_trading["계약월"] = apt_trading["계약년월"] % 100
# apt_trading.head(3) # 앞부분 3줄
# apt_trading.tail(3) # 뒷부분 3줄
```

```python
# 데이터프레임의 구조 출력
apt_trading.info()
```

```python
"a b c".split()[1]
apt_trading['gu'] = apt_trading["시군구"].str.split().str[1] # str Accessor
apt_trading['dong'] = apt_trading["시군구"].str.split().str[2] # str Accessor
apt_trading.info()
```

```python
apt_trading.sort_values(
    by=["거래금액(만원)","전용면적"] , 
    ascending=[True, False]).tail()
```

```python
# GroupBy : XX별 그룹만들기
# 칼럼내 값의 중복값 재베한 값 추출
print(apt_trading["gu"].unique())
print(apt_trading["gu"].nunique())
print(apt_trading["gu"].value_counts()) # 값별 빈도수
```

```python
# 그룹별 통계
# df.groupby("기준칼럼명")[["집계칼럼명"]].통계함수()
apt_gu = apt_trading.groupby("gu")  #구별 거래금액 평균
apt_gu[["거래금액(만원)","pyung"]].mean().sort_values("거래금액(만원)", ascending=False)
```

```python
# 연속형 데이터 : 실수값
# 범주형 데이터 : 종류
# 연속형 데이터 범주형으로 변환 -> 통계
# pd.cut(
#     처리대상,
#     bins=[구간값모음],
#     labels=[구간별이름모음],
#     right=False # 오른쪽 구간값의 포함여부(좌우경계중 하나만 포함)
# )
apt_trading["pyung_type"] = pd.cut(
    apt_trading["pyung"],  #처리대상. 연속형
    bins = [0, 10, 20, 30, 40, 50, 200],
    labels = ["10평미만","10평대","20평대","30평대","40평대","50평이상"],
    right=False
)
```

```python
# 평형별 평균 거래금액
apt_trading.groupby("pyung_type")[["거래금액(만원)"]].mean()
```

```python
# 구별 평형별 평균 거래금액
apt_trading.groupby(["gu","pyung_type"])[["거래금액(만원)"]].mean()
```

```python
#  pivot_table : 표형식 그룹별 통계 계산
# df.pivot_table(
#   index = "주칼럼명", # 선택 - 좌측에 표시
#     columns = "부칼럼명",#선택 - 상단에 표시
#     values="집계칼럼명", # 필수
#     aggfunc="통계함수(기본은 mean)"
# )
round(
    apt_trading.pivot_table(
        values="거래금액(만원)",
        index="pyung_type" ,
        columns="gu",
    ), 2
)
```
