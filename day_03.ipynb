```python
import numpy as np
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
ti = pd.read_csv( url )
ti
```

```python
# ti.isnull()  모든 값별 결측치 여부 반환. 결측치 - True, 정상치 - False
ti.isnull().sum() #칼럼별 결측치 갯수
# ti.isnull().sum().sum() #칼럼별 결측치 갯수를 다 더한것 - 전체 결측치 갯수
```

```python
# age가 결측치인 행 추출
bidx = ti["Age"].isnull()
ti[ bidx ]
```

```python
# 결측치 삭제  df.drop( how="any / all" )
# how="any"  결측치가 1개라도 존재하면 삭제 ,
# how="all"  모든 값이 결측치이면 삭제
ti.dropna() # 행삭제, how="any"
```

```python
# 결측치 재우기
ti["Age"] = ti["Age"].fillna( round(ti["Age"].mean(),1) )
```

# 시각화

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Malgun Gothic'

# data 생성
data = [10,20,30,40]
# plot
plt.plot([1,3,5,7], data, "--r") # 데이터만 입력, 인덱스가 X, 데이터는 Y
plt.plot([2,4,6,8], data, "ob") # 데이터만 입력, 인덱스가 X, 데이터는 Y

plt.title('투명도 (alpha) 설정 예제', fontsize=20)

# X축 & Y축 Label 설정
plt.xlabel('X축', fontsize=20)
plt.ylabel('Y축', fontsize=20)

# 그래프를 보여주는 코드
plt.show()
```
